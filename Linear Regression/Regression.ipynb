 ### Fit Linear Regression Model (Training data)

NOTE :  Now you are required to compute values wihtout using foor loop on tarining data
- HINT : Use numpy library for this purpose 

```python
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x 
#Calculate m
m = x_train.shape[0]
#calculate sum of x^i
sumx = np.sum(x_train)
#calculate sum of (x^i)^2
sumx2 = np.square(x_train).sum()
#calculate sum of y^i
sumy = np.sum(y_train)
#calculate sum of dot of x^i and y^i
sumxy = np.sum(np.dot(x_train,y_train))
#Matrix A and B
A = np.array([[m , sumx], [sumx , sumx2]])
B = np.array([[sumy],[sumxy]])
print("Matrix A:" , A)
print("Matrix B:" , B)
```

Expected output

![Linear Model MAtrixes](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Matrixs%20Linear.png?raw=true)


#### Compute Valus of Both Θ`s


```python
#Your code Here
#Calculate the values of 0_node and 0_1 using Matrix Multiplication
#Hint: Use X = A^-1 * B to calculate the values of 0_node and 0_1
#Taking inverse of A
A_inv = np.linalg.inv(A)

# Solve for matrix X
X = np.dot(A_inv, B)

# Print the result
print("Matrix X for linear Model = ", X)
```

```python
# Your code Here
# In order to calculate the pridicted values of y
# Use the formula y = 0_node + 0_1 * x
# _____DO NOT USE FOR LOOP_____
# instead Convert the column of test_x into numpy a 2 dimensional matrix
# then use matrix multiplication to calculate the predicted values of y
#Read test data file
testData = pd.read_csv('testRegression.csv')
#Extracting X column
x_test = np.array(testData['X'])
y_test = np.array(testData['R'])
#Converting it into 2_dimensional
x_Retest = x_test.reshape(-1,1)
#Predicing values of y on test data
Predicted_y = X[0] + np.dot(x_Retest , X[1])
```

### Comment on the results by comparing all of your models

->First Order Model is appropriate when the relationship between independent and dependent variables is roughly linear
->Second Order Model is appropriate when the relationship between the independent and dependent variables is non-linear and has a parabolic shape.
->Third Order Model is appropriate when the relationship between the independent and dependent variables is non-linear and has a curved shape.
->As we increase the degree of the polynomial, the model becomes more flexible and can fit more complex relationships between the independent and dependent variables.However, this can also lead to overfitting.
